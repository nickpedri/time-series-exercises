## 1. Using the code from the lesson as a guide and the REST API from https://swapi.dev/ as we did in the lesson, create a dataframe named people that has all of the data for people.

```python
def scrape_data(url='link'):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data['results'])
    page = data.copy()
    while page['next'] != None:
        page = requests.get(page['next']).json()
        df = pd.concat([df, pd.DataFrame(page['results'])], ignore_index=True)
    return df
```

```python
people = scrape_data('https://swapi.dev/api/people')
people.head(3)
```
